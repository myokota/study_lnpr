```python
import sys
sys.path.append("../scripts/")
from robot import *
```

```python
world = World(40.0, 0.1)

initial_pose = np.array([0, 0, 0]).T
robots = []

for i in range(10):
    r = Robot(initial_pose, sensor=None, agent=Agent(0.2, 10.0/180*math.pi))
    world.append(r)
    robots.append(r)

r = IdealRobot(initial_pose, sensor=None, agent=Agent(0.2, 10.0/180*math.pi), color='red')
world.append(r)
robots.append(r)

world.draw()
```

```python
import pandas as pd
poses = pd.DataFrame([[math.sqrt(r.pose[0]**2 + r.pose[1]**2), r.pose[2]] for r in robots],
                     columns=['r', 'theta'])
poses.transpose()                      
```

```python
theta_var = poses["theta"].var()
theta_mean = poses["theta"].mean()
theta_per_theta_var = theta_var/theta_mean
std_ww =  math.sqrt(theta_per_theta_var)
print("Θの分散 = ", theta_var)
print("Θの平均 = ", theta_mean)
print("1移動あたりΘの分散 = ", theta_per_theta_var)
print("標準偏差σωω = ", std_ww)
```
