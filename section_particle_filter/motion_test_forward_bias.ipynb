```python
import sys
sys.path.append("../scripts/")
from robot import *
```

```python
world = World(40.0, 0.1)

initial_pose = np.array([0, 0, 0]).T
robots = []

for i in range(10):
    r = Robot(initial_pose, sensor=None, agent=Agent(0.1, 0.0))
    world.append(r)
    robots.append(r)

r = IdealRobot(initial_pose, sensor=None, agent=Agent(0.1, 0.0), color='red')
world.append(r)
robots.append(r)

world.draw()
```

```python
import pandas as pd
poses = pd.DataFrame([[math.sqrt(r.pose[0]**2 + r.pose[1]**2), r.pose[2]] for r in robots],
                     columns=['r', 'theta'])
poses.transpose()                      
```

```python
r_var = poses["r"].var()
r_mean = poses["r"].mean()
r_per_r_var = r_var/r_mean
std_vv =  math.sqrt(r_per_r_var)
print("rの分散 = ", r_var)
print("rの平均 = ", r_mean)
print("1移動あたりrの分散 = ", r_per_r_var)
print("標準偏差σνν = ", std_vv)
```
